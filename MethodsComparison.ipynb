```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_ECT_XGBoost = pd.read_csv('results/ECT+XGBoost.csv')
df_ECT_XGBoost['Method'] = 'ECT+XGBoost'
df_ECT_XGBoost['Category'] = 'ECT'

df_Fingerprint = pd.read_csv('results/Fingerprint+XGBoost.csv')
df_Fingerprint['Method'] = 'Fingerprint+XGBoost'
df_Fingerprint['Category'] = 'Fingerprint'

df_Descriptor = pd.read_csv('results/Descriptor+XGBoost.csv')
df_Descriptor['Method'] = 'Descriptor+XGBoost'
df_Descriptor['Category'] = 'Descriptor'

df_GAT = pd.read_csv('results/GAT.csv')
df_GAT['Method'] = 'GAT'
df_GAT['Category'] = 'GNN'

df_GCN = pd.read_csv('results/GCN.csv')
df_GCN['Method'] = 'GCN'
df_GCN['Category'] = 'GNN'

df_GNN = pd.read_csv('results/GNNattentive.csv')
df_GNN['Method'] = 'GNNattentive'
df_GNN['Category'] = 'GNN'

df_results = pd.concat([df_ECT_XGBoost, df_Fingerprint, df_Descriptor, df_GAT, df_GCN, df_GNN], ignore_index=True)
df_results_train = df_results[df_results['dataset']=="train"].copy()
df_results_tests = df_results[df_results['dataset']=="test"].copy()

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

sns.boxplot(data=df_results_train[df_results_train["metric"]=="RMSE"], x="value",y="Method", hue="Category", ax=axes[0,0])
axes[0,0].set_xlabel('RMSE')
axes[0,0].set_title('RMSE train error per method for ADRA1A dataset')

sns.boxplot(data=df_results_train[df_results_train["metric"]=="R2"], x="value",y="Method", hue="Category", ax=axes[0,1])
axes[0,1].set_xlabel('R2')
axes[0,1].set_title('R2 train error per method for ADRA1A dataset')

sns.boxplot(data=df_results_train[df_results_train["metric"]=="MAE"], x="value",y="Method", hue="Category", ax=axes[0,2])
axes[0,2].set_xlabel('MAE')
axes[0,2].set_title('MAE train error per method for ADRA1A dataset')

sns.boxplot(data=df_results_tests[df_results_tests["metric"]=="RMSE"], x="value",y="Method", hue="Category", ax=axes[1,0])
axes[1,0].set_xlabel('RMSE')
axes[1,0].set_title('RMSE test error per method for ADRA1A dataset')

sns.boxplot(data=df_results_tests[df_results_tests["metric"]=="R2"], x="value",y="Method", hue="Category", ax=axes[1,1])
axes[1,1].set_xlabel('R2')
axes[1,1].set_title('R2 test error per method for ADRA1A dataset')

sns.boxplot(data=df_results_tests[df_results_tests["metric"]=="MAE"], x="value",y="Method", hue="Category", ax=axes[1,2])
axes[1,2].set_xlabel('MAE')
axes[1,2].set_title('MAE test error per method for ADRA1A dataset')

plt.tight_layout()
plt.savefig('figures/boxplotsComparison_ADRA1A.png', dpi=300, bbox_inches='tight')
plt.show()
```
